```python
#Divide the data into train,test and development 60% train, 20% test and 20% development
from sklearn.model_selection import train_test_split

train_size=0.6
# In the first step we will split the data in training and remaining dataset
train,rem = train_test_split(df, train_size=0.6)

test_size = 0.5
dev,test = train_test_split(rem, test_size=0.5)

print(train.count())
print(dev.count())
print(test.count())
```

```python
#list to store all reviews of the train data
col_list = train['Review'].tolist()
```

```python
#store all words in the dictionary and their store frequency
import re
word_count = {}
for row in col_list:
    words = list(set(re.split("[!. _,@?:-= ]", row.lower())))
    words = [i for i in words if i !='']
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
#print ten words of the dictionary
first_ten = list(word_count.items())[:10]
print(first_ten)  
```

```python
print(train['Freshness'])
```

```python
#lists to store all reviews which are fresh and rotten separately

freshData = []
rottenData = []

for i, row in train.iterrows():
    if row['Freshness'] == "fresh":
        freshData.append(row['Review'])
    if row['Freshness'] == "rotten":
        rottenData.append(row['Review'])

freshData = [x.lower() for x in freshData]
rottenData = [x.lower() for x in rottenData]
    
```

```python
print(freshData[:10])
```

```python
#store all words of fresh in the dictionary and their store frequency
import re
fresh_count = {}
for row in freshData:
    words = list(set(re.split("[!. _,@?:-=]", row.lower())))
    words = [i for i in words if i !='']
    for word in words:
        if word in fresh_count:
            fresh_count[word] += 1
        else:
            fresh_count[word] = 1
#print ten words of the dictionary
first_ten = list(fresh_count.items())[:10]
print(first_ten)
```

```python
#store all words of rotten in the dictionary and their store frequency
import re
rotten_count = {}
for row in rottenData:
    words = list(set(re.split("[!. _,@?:-=]", row.lower())))
    words = [i for i in words if i !='']
    for word in words:
        if word in rotten_count:
            rotten_count[word] += 1
        else:
            rotten_count[word] = 1
#print ten words of the dictionary
first_ten = list(rotten_count.items())[:10]
print(first_ten)
```

```python
#Count number of rows in the train dataset
total_count=train['Review'].count()
#count number of rows which are rotten
count_rotten=len(rottenData)
#count number of rows which are fresh
count_fresh=len(freshData)

print(total_count, count_rotten, count_fresh)
```

```python
prob_rotten=count_rotten/total_count
prob_fresh=count_fresh/total_count

print(prob_rotten,prob_fresh)
```

```python
def predict_class(review):
    a=re.split("[!. _,@?:-= ]",review.lower())
    a = [i for i in a if i !='']
    ans_rotten=1
    ans_fresh=1
    for ele in a:
        ans_rotten=ans_rotten*rotten_count.get(ele, 0)/count_rotten
        ans_fresh=ans_fresh*fresh_count.get(ele, 0)/count_fresh
    fresh_prob=(ans_fresh*prob_fresh)/(ans_fresh*prob_fresh+ ans_rotten*prob_rotten) 
    rotten_prob=(ans_rotten*prob_rotten)/(ans_fresh*prob_fresh+ ans_rotten*prob_rotten) 
    if(fresh_prob>rotten_prob):
        return "fresh"
    else:
        return "rotten"
```

```python
pred=[]
for i,row in dev.iterrows():
    pred.append(predict_class(row['Review']))
```

```python
total=len(pred)
count=0
for i in range(total):
    if pred[i]==d[i]:
        count=count+1
accuracy = count/total
print(f"Accuracy: {accuracy:.2f}")
```

```python
#Smoothing
def predict_class_smooth(review):
    a=re.split("[!. _,@?:-= ]",review.lower())
    a = [i for i in a if i !='']
    ans_rotten=1
    ans_fresh=1
    for ele in a:
        if(ele in rotten_count):
            ans_rotten*=(rotten_count[ele]+1)/(count_rotten+1*2)
        if(ele not in rotten_count):
            ans_rotten*=(1)/(count_rotten+1*2)
        if(ele in fresh_count):
            ans_fresh=(fresh_count[ele]+1)/(count_fresh+1*2)
        if(ele not in fresh_count):
            ans_fresh*=(1)/(count_fresh+1*2)
    fresh_prob=(ans_fresh*prob_fresh)/(ans_fresh*prob_fresh+ ans_rotten*prob_rotten) 
    rotten_prob=(ans_rotten*prob_rotten)/(ans_fresh*prob_fresh+ ans_rotten*prob_rotten) 
    if(fresh_prob>rotten_prob):
        return "fresh"
    else:
        return "rotten"
```

```python
pred_smooth=[]
for i,row in dev.iterrows():
    pred_smooth.append(predict_class_smooth(row['Review']))
```

```python
#calculate new accuracy
total=len(pred_smooth)
count=0
for i in range(total):
    if pred_smooth[i]==d[i]:
        count=count+1
accuracy = count/total
print(f"Accuracy: {accuracy:.2f}")
```
